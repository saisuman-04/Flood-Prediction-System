```python
plt.figure(figsize=(15,12))

sns.heatmap(df.corr(), cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["FloodProbability"], bins=30, kde=True)

plt.title("Distribution of Flood Probability")

plt.show()
```

```python
df["Flood"] = (df["FloodProbability"] >= 0.5).astype(int)

df[["FloodProbability", "Flood"]].head()
```

```python
df["Flood"].value_counts()
plt.figure(figsize=(6,4))

sns.countplot(x="Flood", data=df)

plt.title("Flood vs No Flood")

plt.show()
```
